```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy
from tensorflow.keras.activations import sigmoid
```

```python
x_train = np.array([[1.0],[2.0]], dtype=np.float32)
y_train = np.array([[300.0],[500.0]], dtype=np.float32)

# visualise the data with scatterplot, add labes as y price in 1000s of dollars, x size in square feet
plt.scatter(x_train, y_train)
plt.xlabel('Size (Square Feet)')
plt.ylabel('Price (1000s of $)')
plt.show()

```

```python
linear_layer = tf.keras.layers.Dense(units=1, activation='linear')
#The line of code linear_layer = tf.keras.layers.Dense(units=1, activation='linear') is creating 
#a dense (fully connected - each neuron in the layer is connected to every neuron in the previous layer) layer 
#using TensorFlow's Keras API. This layer is a fundamental building block in neural networks, particularly in deep learning models.

#The activation function determines how the output of the layer is transformed. A linear activation function means that the output is a linear combination of the inputs, 
# without any non-linear transformation. 
#This is essentially the identity function, f(x) = x, and is often used in the output layer of regression models where the goal is to predict a continuous value.
```

```python
x = np.array([[200.0, 17.0],
            [120.0, 5.0],
            [425.0, 20.0],
            [212.0, 18.0]])

y = np.array([1,0,0,1])

layer_1 = Dense(units=3, activation='sigmoid')
layer_2 = Dense(units=1, activation='sigmoid')
model = Sequential([layer_1, layer_2])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x,y, epochs=1000)
```

```python
model.predict(np.array([[200.0, 17.0]]))
```
