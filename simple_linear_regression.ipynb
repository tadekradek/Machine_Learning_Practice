```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Simulate 1000 samples of house sizes (in square feet)
house_sizes = np.random.normal(loc=1500, scale=500, size=1000)

# Assume a linear relationship between house size and price
# Price = 100 * house_size + 50,000 + some noise
noise = np.random.normal(loc=0, scale=20000, size=1000)
house_prices = 100 * house_sizes + 50000 + noise

# Create a DataFrame
data = pd.DataFrame({'House Size (sq ft)': house_sizes, 'House Price (USD)': house_prices})

# Display the first few rows of the DataFrame
data.head()

#rename columns for the following:  'HouseSize' and 'HousePrice'
data.columns = ['HouseSize', 'HousePrice']

```

```python
# plot the data
import matplotlib.pyplot as plt
plt.scatter(data['HouseSize'], data['HousePrice'])
plt.xlabel('House Size (sq ft)')
plt.ylabel('House Price (USD)')
plt.title('House Price vs House Size')
plt.show()

```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data' DataFrame is already defined with columns 'HouseSize' and 'HousePrice'
# Extract features and target
X = data['HouseSize'].values.reshape(-1, 1)
y = data['HousePrice'].values.reshape(-1, 1)

# Add a column of ones to X to account for the bias term
X_b = np.c_[np.ones((X.shape[0], 1)), X]

# Initialize parameters
theta = np.array([6, 0.3]).reshape(-1, 1)  # [b, theta]
alpha = 0.001
iterations = 3000
m = len(y)

# Gradient Descent
costs = []
params = []

for iteration in range(iterations):
    predictions = X_b.dot(theta)
    errors = predictions - y
    cost = (1/(2*m)) * np.sum(errors**2)
    gradients = (1/m) * X_b.T.dot(errors)
    theta -= alpha * gradients
    
    # Store the cost and parameters
    costs.append(cost)
    params.append(theta.flatten())

# Create a DataFrame to store the results
results_df = pd.DataFrame(params, columns=['b', 'theta'])
results_df['Cost'] = costs




```
